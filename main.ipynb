```python
import pandas as pd
import numpy as np
from scipy import stats as stats
import matplotlib.pyplot as plt
```

```python
#Lưu và chuyển dữ liệu đã xử lý vào biến data và lấy biến giá
data = pd.read_csv('Intel_CPUs_Cleaned_Final.csv')
prices = data['Recommended_Customer_Price']
```

```python
n = len(prices)
x_bar = prices.mean()
s = prices.std(ddof=1)  # ddof=1 để tính độ lệch chuẩn mẫu
```

```python
# Vẽ Q-Q Plot
plt.figure(figsize=(6, 4))
stats.probplot(prices, dist="norm", plot=plt)
plt.title("Q-Q Plot of Recommended_Customer_Price")
plt.show()
```

```python
# Kiểm định Shapiro-Wilk
stat, p_value_shapiro = stats.shapiro(prices)
print(f"Shapiro-Wilk Test: Statistic={stat:.4f}, p-value={p_value_shapiro}")
```

```python
# Thực hiện One-Sample T-test
t_stat, p_val = stats.ttest_1samp(a=prices, popmean=600)
print(f"Giá trị t: {t_stat:.4f}, p-value: {p_val:.4f}")

# Tính khoảng tin cậy 95%
ci_lower, ci_upper = stats.t.interval(confidence=0.95, df=n-1, 
                                      loc=x_bar, scale=s/np.sqrt(n))
print(f"Khoảng tin cậy 95%: [{ci_lower:.2f}, {ci_upper:.2f}]")
```
